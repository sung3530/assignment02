# Import packages for graphs and data

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define my function : $f(x) = \sin(x) \cdot x$

```python
def myFuction(x):
    f=np.sin(x)*x
    return f
```

# Define the derivative of my function : $f'(x) = \cos(x) \cdot x + \sin(x)$ 

```python
def myDerivativeFunction(x):
    Df=np.cos(x) * x + np.sin(x)
    return Df
```

# Define the domain of the function: x = [-20:0.1:20]

```python
x = np.arange(-20,20,0.1)
```

# Compute the graph

```python
f=myFuction(x)
Df=myDerivativeFunction(x)
```

# Plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x,f,'b',label="function")
plt.plot(x,Df,'r',label="derivative")
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0)
plt.show()
```

# Pick 3 points in the domain

```python
x1 = -3
x2 = 4
x3 = 2
```

# First-order Taylor approximation $g(x) = f(z) + f'(z)(x-z)$


```python
def myTaylorFunction(z,x):
    tf=myFuction(z)+myDerivativeFunction(z)*(x-z)
    return tf
```

# Compute the Taylor approximation function

```python
tf1=myTaylorFunction(x1,x)
tf2=myTaylorFunction(x2,x)
tf3=myTaylorFunction(x3,x)

```

# Plot the graphs for the function of the Taylor approximation

```python
plt.figure(1)
plt.plot(x,f,'b',label="function")
plt.plot(x,tf1,'r',label="derivative")
plt.plot(x,tf2,'g',label="derivative")
plt.plot(x,tf3,'c',label="derivative")
plt.show()
```
